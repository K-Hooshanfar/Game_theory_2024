```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def replicator_dynamics(payoff_matrix, population, timesteps):
    num_strategies = len(population)
    x_history = [population[0]]
    y_history = [population[1]]
    z_history = [population[2]]

    for xxx in range(timesteps):
        fitness = np.dot(payoff_matrix, population)
        average_fitness = np.dot(population, fitness)
        for i in range(num_strategies):
            new_population = population[i] * (fitness[i] - average_fitness)
            population[i] += new_population



        x_history.append(population[0])
        y_history.append(population[1])
        z_history.append(population[2])
        print(xxx,population)

    # Plot 3D population evolution
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    ax.plot(x_history, y_history, z_history, marker='o')
    ax.set_xlabel('Population[0]')
    ax.set_ylabel('Population[1]')
    ax.set_zlabel('Population[2]')
    plt.show()
    print(population)
    return population
```

```python
payoff_matrix = np.array([[1, 3, 2], [0, 2, 5], [6, 1, 4]])

#initial_population = np.array([1/3, 1/3, 1/3])
initial_population = np.array([1/7, 17/28, 1/4])
#final_population = replicator_dynamics(payoff_matrix, initial_population,30)
population =  replicator_dynamics(payoff_matrix, initial_population,20)


```
